```python
from mwprojekt.traffic import traffic_computation

traffic = traffic_computation('../data/','03.02.2021',4)

traffic = [x[x['kph']>20] for x in traffic]
```

```python
#wyszukuje obszary, w których autobusy przekraczają 20km/h
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


for x_df in traffic:
    kmeans = KMeans(n_clusters=2).fit(x_df)
    centroids = kmeans.cluster_centers_

    plt.scatter(x_df['Lat_y'],x_df['Lon_y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
    plt.xlabel("longitude y")
    plt.ylabel("latitude x")
    plt.show()
```
